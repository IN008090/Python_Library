### How to import file from csv and excel files

```python
import pandas as pd
stockmarket_dataset={  # Create DataFrame using list within Dict 
    'Tickers':['GOOGL','WMT','MSFT','RIL','TATA'],
    'EPS':['27.82','4.61','-1','NA','5.6'],
    'REVENUE':['87','484','85','50','-1'],
    'PRICE':['845','65','64','1023','NA'],
    'PEOPLE':['Larry page','NA','Bill Gate','Mukesh Ambani','Ratan Tata']
}   # Methord 1 to import data using Dict
df=pd.DataFrame(stockmarket_dataset)
df
```
