### Topic covered
+ fillna to fill missing values using different ways.
+ interpolate to make a guess on missing values using interpolation
+ dropna to drop rows with missing values.

```python
# How to import the csv file into Notebook
import pandas as pd 
df=pd.read_csv('Missing_dataset_01.csv',parse_dates=['Date'])  # To convert the date format use parse_dates=['Name of the date column']
df
```

```python
type(df.Date[0])  # This shows the type in dataset
```

```python
df.set_index('Date',inplace=True)  # This is used to convert the Particular column into Index and mandatory to use 'True'
```

```python
new_df=df.fillna(0) # To fill the empty dataset or null values 'df.fillna(followed by what you want to change)'
#If you want to do somechanges in the dataset you have two option either you do the changes in the original dataset or make it new dataset by just name it  
# Here we defined 'new_df' as new dataset it will not disturb the original dataset 'df'
```

```python
new_df
```

```python
new_df2=df.fillna({
    'Temperature':'0',
    'Windspeed':'0',
    'Event':'no event'
})
new_df2   # If want to modify all the 'NA' values but  different values on different column
```

```python
new_df=df.fillna(method='ffill') # To carry forwards the last value inplace of zero to  make the data meaningful
new_df
```

```python
new_df=df.fillna(method='bfill') # To fill the NA data in backward 
new_df
```

```python
new_df=df.fillna(method="bfill",axis='columns')  # To fill along the column
new_df
```

```python
new_df=df.fillna(method='bfill',limit=1)# To limit how many value it need to copied and remainig value will be 'NA'
new_df
```

```python
new_df=df.interpolate()  # This show the meaning graphically automatically without any guesses
new_df
```

```python
new_df=df.interpolate(methord='time') # Check in the interpolate in doc. different types method
new_df
```

```python
new_df=df.dropna()  # To drop the NA values
new_df
```

```python
new_df=df.dropna(how='all') # If all cells have NA it will drop rest if have value will remain as it
new_df
```

```python
dt=pd.date_range("01-01-2017","01-11-2017")
idx=pd.DatetimeIndex(dt)
df=df.reindex(idx)   # To get remaining date in between as in the dataset from date 1 its 4 so remaining date
df
```

```python
new_df=df.dropna(thresh=1) # Atleast one valid value other then NA 
new_df
```
