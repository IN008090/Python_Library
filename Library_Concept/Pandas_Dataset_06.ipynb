```python
import pandas as pd
import numpy as np # np is the numpy 
df=pd.read_csv('New_df01.csv')
df
```

```python
new_df=df.replace(-99999,np.NaN) #To replace the special value with NA we use numpy lib.
new_df  # Note in numpy only NaN is used can not use another word
```

```python
new_df=df.replace([-99999,-88888],np.NaN)
new_df
```

```python
new_df=df.replace({
    -99999:'NaN',
    -88888:'NaN'   # Second methord to replace the special no with dict
})
new_df
```

```python
new_df=df.replace({
    'temperature':-99999,
    'windspeed':-88888,
    'event':"No Event"  # If need to replace the special value within a particular column can use dict 
},np.NaN)
new_df
```

```python
new_df 
new_df=df.replace({
-99999:np.NaN,   # Another methord to replace the special no
'No Event':'sunny'
})
new_df
```

```python
new_df=df.replace('[A-Za-z]','',regex=True)#regex tutorial very import
new_df
# This will remove the the alphabet which are along the number from each row but if some row have alphbet data it will replace that too
```

```python
new_df=df.replace({
'temperature':'[A-Za-z]',
'windspeed':'[A-Za-z]'
},'',regex=True)
new_df   # While using the dict we can choose from each row which need to be select.
```

```python
df=pd.DataFrame({    # Make sure if you defined the new original dataset name it as df as 
    'Score':['exceptional','average','good','poor','average','exceptional'],
    'Student':['rob','maya','parthiv','tom','julian','erica']
})
df
```

```python
new_df=df.replace(['average','good','poor','average','exceptional'],['B','C','D','E','F'])  # With list method
new_df
```

```python
new_df=df.replace({  # With dict method
    'average':'B',
    'exceptional':'F',
    'good':'A',
    'poor':'C'
})
new_df
```
